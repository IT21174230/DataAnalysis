```python
dataset='/content/drive/MyDrive/waterquality/waterQuality1.csv'

import pandas as pd
dataframe=pd.read_csv(dataset)

data=dataframe.head()
data

print(dataframe.dtypes)

dataframe.isnull().sum()
```

```python
dataframe.replace('#NUM!', pd.NA, inplace=True)
dataframe.dropna(inplace=True)
dataframe.reset_index(drop=True, inplace=True)

print(dataframe)
```

```python
dataframe.drop(columns=['is_safe'])
```

```python
correlation_mtrx=dataframe.corr()
```

```python
print(correlation_mtrx[:5][:5])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_mtrx, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
correlation_threshold=0.5

def find_pairs(threshold, mtrx):
  pairs=[]
  for i in range(len(mtrx.columns)):
    for j in range(i):
      if abs(mtrx.iloc[i, j]) > correlation_threshold:
        pairs.append([mtrx.columns[i], mtrx.columns[j]])
  return pairs
cor_pairs=find_pairs(correlation_threshold, correlation_mtrx)
print(cor_pairs)
```

```python

for i in cor_pairs:
    if correlation_mtrx.loc['uranium', i[0]] > correlation_mtrx.loc['uranium', i[1]]:
        if i[1] in dataframe:
            dataframe = dataframe.drop(columns=[i[1]])
    elif i[0] in dataframe:
        dataframe = dataframe.drop(columns=[i[0]])

dataframe.head()
```
